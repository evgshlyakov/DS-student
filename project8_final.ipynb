```python
# расчёт эффективности обогащения чернового концентрата, сравнение с исходными данными

gold_train['rougher.recovery_calc'] = gold_train['rougher.output.concentrate_au'] * (
    gold_train['rougher.input.feed_au'] - gold_train['rougher.output.tail_au'])

gold_train['rougher.recovery_calc'] = gold_train['rougher.recovery_calc'] / (
    gold_train['rougher.input.feed_au'] * (gold_train['rougher.output.concentrate_au'] - 
                                          gold_train['rougher.output.tail_au'])) * 100

# вывод исходных данных, исходноой и рассчётной эффективности обогащения чернового концентрата 
gold_train[['rougher.input.feed_au', 'rougher.output.tail_au', 'rougher.output.concentrate_au', 
           'rougher.output.recovery', 'rougher.recovery_calc']]
```

```python
# функции расчёта метрики sMAPE классического и интегрированного sMAPE

def smape(target, predicted):
    """
    Принимает фактический и предсказанный массив значений, а возвращает классическую метрику sMAPE в %
    """
    return 100/len(target) * np.sum(2 * np.abs(target - predicted) / (np.abs(target) + np.abs(predicted)))


def smape_total(rougher, rougher_predicted, final, final_predicted):
    """
    Принимает фактические значения эффективности обогащения чернового и финального
    концентратов, их предсказания, а возвращает метрику sMAPE в %, специфическую для данного исследования
    """
    return 0.25 * smape(rougher, rougher_predicted) + 0.75 * smape(final, final_predicted)
```

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'> `smape` и `smape_total` расчитаны верно.</font>

```python
# формируем скорер классического sMAPE для использования в кросс-валидации

smape_scorer = make_scorer(smape, greater_is_better=False)
```

<font color='green'>Верно, что не забываешь указать greater_is_better=False, Это пригодится для SearchCV. SMAPE — ошибка, её минимизируем.</font>

*Значение итогового sMAPE, в размере 25% от sMAPE флотации плюс 75% от sMAPE всего производства, позволяет предсказывать коэффициенты эффективности обогащения чернового и финального концентратов с отклонением в 9.5%., что не превышает порог адекватности модели (9.83%), если бы предсказания велись по среднему значению целевых признаков.*
