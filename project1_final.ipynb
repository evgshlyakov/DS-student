**–ó–∞–¥–∞–Ω–∏–µ 4. –í—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –¥–≤—É—Ö –º–µ—Ç–æ–¥–æ–≤.**

```python
data.isna().sum()
```

**–ó–∞–¥–∞–Ω–∏–µ 5. –í –¥–≤—É—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö –µ—Å—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. –û–¥–∏–Ω –∏–∑ –Ω–∏—Ö ‚Äî `days_employed`. –ü—Ä–æ–ø—É—Å–∫–∏ –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –≤—ã –æ–±—Ä–∞–±–æ—Ç–∞–µ—Ç–µ –Ω–∞ —Å–ª–µ–¥—É—é—â–µ–º —ç—Ç–∞–ø–µ. –î—Ä—É–≥–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ ‚Äî `total_income` ‚Äî —Ö—Ä–∞–Ω–∏—Ç –¥–∞–Ω–Ω—ã–µ –æ –¥–æ—Ö–æ–¥–∞—Ö. –ù–∞ —Å—É–º–º—É –¥–æ—Ö–æ–¥–∞ —Å–∏–ª—å–Ω–µ–µ –≤—Å–µ–≥–æ –≤–ª–∏—è–µ—Ç —Ç–∏–ø –∑–∞–Ω—è—Ç–æ—Å—Ç–∏, –ø–æ—ç—Ç–æ–º—É –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –Ω—É–∂–Ω–æ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–∏–ø—É –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `income_type`. –ù–∞–ø—Ä–∏–º–µ—Ä, —É —á–µ–ª–æ–≤–µ–∫–∞ —Å —Ç–∏–ø–æ–º –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ `—Å–æ—Ç—Ä—É–¥–Ω–∏–∫` –ø—Ä–æ–ø—É—Å–∫ –≤ —Å—Ç–æ–ª–±—Ü–µ `total_income` –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∑–∞–ø–æ–ª–Ω–µ–Ω –º–µ–¥–∏–∞–Ω–Ω—ã–º –¥–æ—Ö–æ–¥–æ–º —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –∑–∞–ø–∏—Å–µ–π —Å —Ç–µ–º –∂–µ —Ç–∏–ø–æ–º.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

**–ó–∞–¥–∞–Ω–∏–µ 6. –í –¥–∞–Ω–Ω—ã—Ö –º–æ–≥—É—Ç –≤—Å—Ç—Ä–µ—á–∞—Ç—å—Å—è –∞—Ä—Ç–µ—Ñ–∞–∫—Ç—ã (–∞–Ω–æ–º–∞–ª–∏–∏) ‚Äî –∑–Ω–∞—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –æ—Ç—Ä–∞–∂–∞—é—Ç –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏ –ø–æ—è–≤–∏–ª–∏—Å—å –ø–æ –∫–∞–∫–æ–π-—Ç–æ –æ—à–∏–±–∫–µ. —Ç–∞–∫–∏–º –∞—Ä—Ç–µ—Ñ–∞–∫—Ç–æ–º –±—É–¥–µ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π —Ç—Ä—É–¥–æ–≤–æ–≥–æ —Å—Ç–∞–∂–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ `days_employed`. –î–ª—è —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —ç—Ç–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ. –û–±—Ä–∞–±–æ—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ: –∑–∞–º–µ–Ω–∏—Ç–µ –≤—Å–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º–∏ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ `abs()`.**

```python
data['days_employed'] = data['days_employed'].abs()
```

### –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

**–ó–∞–¥–∞–Ω–∏–µ 16. –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤, —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –Ω–∏–∂–µ, —Å–æ–∑–¥–∞–π—Ç–µ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ `data` —Å—Ç–æ–ª–±–µ—Ü `total_income_category` —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏:**

- 0‚Äì30000 ‚Äî `'E'`;
- 30001‚Äì50000 ‚Äî `'D'`;
- 50001‚Äì200000 ‚Äî `'C'`;
- 200001‚Äì1000000 ‚Äî `'B'`;
- 1000001 –∏ –≤—ã—à–µ ‚Äî `'A'`.


**–ù–∞–ø—Ä–∏–º–µ—Ä, –∫—Ä–µ–¥–∏—Ç–æ–ø–æ–ª—É—á–∞—Ç–µ–ª—é —Å –¥–æ—Ö–æ–¥–æ–º 25000 –Ω—É–∂–Ω–æ –Ω–∞–∑–Ω–∞—á–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏—é `'E'`, –∞ –∫–ª–∏–µ–Ω—Ç—É, –ø–æ–ª—É—á–∞—é—â–µ–º—É 235000, ‚Äî `'B'`. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å –∏–º–µ–Ω–µ–º `categorize_income()` –∏ –º–µ—Ç–æ–¥ `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

- `'–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'`,
- `'–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'`,
- `'–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'`,
- `'–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'`.

**–ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ `purpose` –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ `'–Ω–∞ –ø–æ–∫—É–ø–∫—É –∞–≤—Ç–æ–º–æ–±–∏–ª—è'`, —Ç–æ –≤ —Å—Ç–æ–ª–±—Ü–µ `purpose_category` –¥–æ–ª–∂–Ω–∞ –ø–æ—è–≤–∏—Ç—å—Å—è —Å—Ç—Ä–æ–∫–∞ `'–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'`.**

**–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å –∏–º–µ–Ω–µ–º `categorize_purpose()` –∏ –º–µ—Ç–æ–¥ `apply()`. –ò–∑—É—á–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ `purpose` –∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –∫–∞–∫–∏–µ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ –ø–æ–º–æ–≥—É—Ç –≤–∞–º –ø—Ä–∞–≤–∏–ª—å–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏—é.**

```python
def categorize_purpose(row):
    try:
        if '–∞–≤—Ç–æ–º' in row:
            return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'
        elif '–∂–∏–ª' in row or '–Ω–µ–¥–≤–∏–∂' in row:
            return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'
        elif '—Å–≤–∞–¥' in row:
            return '–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'
        elif '–æ–±—Ä–∞–∑–æ–≤' in row:
            return '–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'
    except:
        return '–Ω–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∏'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

## –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –∏ –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã

**–ó–∞–¥–∞–Ω–∏–µ 19. –ï—Å—Ç—å –ª–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ç–µ–π –∏ –≤–æ–∑–≤—Ä–∞—Ç–æ–º –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫?**

   *–î–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–µ—Ç–µ–π –∏ –≤–æ–∑–≤—Ä–∞—Ç–∞ –∫—Ä–µ–¥–∏—Ç–∞ —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É –∏–∑ —Å–ª–µ–¥—É—é—â–∏—Ö –∫–æ–ª–æ–Ω–æ–∫ —Ç–∞–±–ª–∏—Ü—ã `'data`', –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–º –ø–æ—Ç—Ä–µ–±—É—é—Ç—Å—è –¥–ª—è –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å –∑–∞–¥–∞–Ω–∏—è:*
    
   * `'children`'

  *–¢–∞–∫–∂–µ –≤–æ–∑—å–º—ë–º –∏–∑ —Ç–∞–±–ª–∏—Ü—ã `'data`' –µ—â—ë –æ–¥–Ω—É –ª—é–±—É—é –∫–æ–ª–æ–Ω–∫—É, –≤ –∫–æ—Ç–æ—Ä—É—é –ø–æ–∑–∂–µ –∑–∞–ø–∏—à–µ–º –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –¥–æ–ª–≥–∞ —Å–ª–æ–≤–∞–º–∏, –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–µ–π –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –ø—Ä–∏ –≤—ã–≤–æ–¥–µ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–æ–∫ –∏ —Å–≤–æ–¥–Ω—ã—Ö —Ç–∞–±–ª–∏—Ü*

   *–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —á–∏—Å–ª–æ –¥–æ–ª–∂–Ω–∏–∫–æ–≤ –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø–æ–¥–∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:*
   
   * –±–µ–∑ –¥–µ—Ç–µ–π
   * –∏–º–µ—é—â–∏–µ 1-2 —Ä–µ–±—ë–Ω–∫–∞
   * –∏–º–µ—é—â–∏–µ 3-—Ö –∏ –±–æ–ª–µ–µ –¥–µ—Ç–µ–π

```python
def children_number(number):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é –ø–æ –Ω–∞–ª–∏—á–∏—é –¥–µ—Ç–µ–π, –∏—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–∞:
    - '–±–µ–∑ –¥–µ—Ç–µ–π', –µ—Å–ª–∏ number == 0;
    - '1-2 —Ä–µ–±—ë–Ω–∫–∞', –µ—Å–ª–∏ number == 0 –∏–ª–∏ number == 0;
    - '–º–Ω–æ–≥–æ–¥–µ—Ç–Ω—ã–µ', –µ—Å–ª–∏ number —Ä–∞–≤–Ω–æ 3 –∏ –±–æ–ª–µ–µ.
    """
    try:
        if number == 0:
            return '–±–µ–∑ –¥–µ—Ç–µ–π'
    
        if number == 1 or number == 2:
            return '1-2 —Ä–µ–±—ë–Ω–∫–∞'
    
        if number >= 3:
            return '–º–Ω–æ–≥–æ–¥–µ—Ç–Ω—ã–µ'
    except: 
        pass
```

*–°–µ–º–µ–π–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ –Ω–∞–ª–∏—á–∏—é –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç–∏, –∞ –æ–¥–∏–Ω–æ—á–Ω—ã–µ –∫–∞–∫ –¥–∏—Å—Ü–∏–ø–ª–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ, —Ç–∞–∫ –∏ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –¥–æ–ª–µ–π –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç–∏. –õ–æ–≥–∏—á–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å–µ–º–µ–π–Ω–æ–≥–æ —Å—Ç–∞—Ç—É—Å–∞ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –¥–≤–µ –ø–æ–¥–∫–∞—Ç–µ–≥–æ—Ä–∏–∏:*

   * —Å–µ–º–µ–π–Ω—ã–µ (–∂–µ–Ω–∞—Ç/–∑–∞–º—É–∂–µ–º –∏ –≥—Ä–∞–∂–¥–∞–Ω—Å–∫–∏–π –±—Ä–∞–∫)
   * –æ–¥–∏–Ω–æ—á–∫–∏ (–æ—Å—Ç–∞–ª—å–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏)


```python
def single_double(status):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é —Å–µ–º–µ–π–Ω–æ–≥–æ –∏–ª–∏ –æ–¥–∏–Ω–æ—á–∫–∏, –∏—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–∞:
    - '—Å–µ–º–µ–π–Ω—ã–µ', –µ—Å–ª–∏ status - '–∂–µ–Ω–∞—Ç / –∑–∞–º—É–∂–µ–º' –∏–ª–∏ '–≥—Ä–∞–∂–¥–∞–Ω—Å–∫–∏–π –±—Ä–∞–∫';
    - '–æ–¥–∏–Ω–æ—á–∫–∏', –µ—Å–ª–∏ status - '–Ω–µ –∂–µ–Ω–∞—Ç / –Ω–µ –∑–∞–º—É–∂–µ–º', '–≤ —Ä–∞–∑–≤–æ–¥–µ' –∏–ª–∏ '–≤–¥–æ–≤–µ—Ü / –≤–¥–æ–≤–∞'
    """
    try:
        if status == '–∂–µ–Ω–∞—Ç / –∑–∞–º—É–∂–µ–º' or status == '–≥—Ä–∞–∂–¥–∞–Ω—Å–∫–∏–π –±—Ä–∞–∫':
            return '—Å–µ–º–µ–π–Ω—ã–µ'
    
        elif status == '–Ω–µ –∂–µ–Ω–∞—Ç / –Ω–µ –∑–∞–º—É–∂–µ–º' or status == '–≤ —Ä–∞–∑–≤–æ–¥–µ' or \
            status == '–≤–¥–æ–≤–µ—Ü / –≤–¥–æ–≤–∞':
            return '–æ–¥–∏–Ω–æ—á–∫–∏'
        else: return '–Ω–µ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω'
    except: 
        return '–Ω–µ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω'
```

```python
# –ø—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é single_double –∫ —Å—Ç–æ–ª–±—Ü—É 'family_status' –Ω–∞—à–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
data_an['family_group'] = data_an['family_status'].apply(single_double)

# —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–∏—Å—Ü–∏–ø–ª–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∏ –¥–æ–ª–∂–Ω–∏–∫–æ–≤, –∏—Å—Ö–æ–¥—è –∏–∑ –∫–∞—Ç–µ–≥–æ—Ä–∏–π —Å–µ–º–µ–π–Ω–æ–≥–æ —Å—Ç–∞—Ç—É—Å–∞
pivot_family_group = data_an.pivot_table(index='family_group', columns='debt_category', values='debt', aggfunc='count')

# —Å–æ–∑–¥–∞–¥–∏–º –∫–æ–ª–æ–Ω–∫—É —Å –ø—Ä–æ—Ü–µ–Ω—Ç–æ–º –¥–æ–ª–∂–Ω–∏–∫–æ–≤ –ø–æ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏, –æ–∫—Ä—É–≥–ª–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
pivot_family_group['debt %'] = pivot_family_group['debt'] / (pivot_family_group['debt'] + pivot_family_group['clear']) * 100
pivot_family_group['debt %'] = round(pivot_family_group['debt %'], 1)
pivot_family_group.sort_values('debt %', ascending=False)
```

*–ò–∑ —Ç–∞–±–ª–∏—Ü—ã –ø–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ –Ω–∏–∑–∫–∏–µ —É—Ä–æ–≤–Ω–∏ –¥–æ—Ö–æ–¥–∞ —É–≤–µ–ª–∏—á–∏–≤–∞—é—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç–∏, –ø—Ä–∏ —ç—Ç–æ–º –∑–∞—ë–º—â–∏–∫–∏ —Å –¥–æ—Ö–æ–¥–∞–º–∏ —á—É—Ç—å –≤—ã—à–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö —Å–∞–º—ã–µ –¥–∏—Å—Ü–∏–ø–ª–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ. –û–¥–Ω–∞–∫–æ, –¥–∞–Ω–Ω–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –Ω–µ —Å–ª–∏—à–∫–æ–º —É–¥–∞—á–Ω–∞, —Å–∏–ª—å–Ω—ã–π –ø–µ—Ä–µ–∫–æ—Å –≤ —Å—Ç–æ—Ä–æ–Ω—É –æ–¥–Ω–æ–π –∏–∑ –≥—Ä—É–ø–ø. –°–¥–µ–ª–∞–µ–º –¥—Ä—É–≥—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—é, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–≤ —Ñ—É–Ω–∫—Ü–∏—é –æ–¥–Ω–æ–π –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –∑–∞–¥–∞—á*

```python
def categorize_income_2(income):
    """
    –†–∞–∑–±–∏–≤–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–æ—Ö–æ–¥–æ–≤ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≥—Ä—É–ø–ø—ã:
    - '–¥–æ 70 000' —Ä—É–±–ª–µ–π,
    - '–æ—Ç 70 001 –¥–æ 100 000' —Ä—É–±–ª–µ–π
    - '–æ—Ç 100 001 –¥–æ 150 000' —Ä—É–±–ª–µ–π
    - '–æ—Ç 150 001 –¥–æ 300 000' —Ä—É–±–ª–µ–π
    - '—Å–≤—ã—à–µ 300 001' —Ä—É–±–ª–µ–π
    """
    try:
        if 0 <= income <= 70000:
            return '–¥–æ 70 000'
        elif 70001 <= income <= 100000:
            return '–æ—Ç 70 001 –¥–æ 100 000'
        elif 100001 <= income <= 150000:
            return '–æ—Ç 100 001 –¥–æ 150 000'
        elif 150001 <= income <= 300000:
            return '–æ—Ç 150 001 –¥–æ 300 000'
        elif income >= 300001:
            return '—Å–≤—ã—à–µ 300 001'
    except:
        pass
```

```python
# —Å–∫–æ–ø–∏—Ä—É–µ–º —Å—Ç–æ–ª–±–µ—Ü 'total_income' –∏–∑ —Ç–∞–±–ª–∏—Ü—ã data
data_an['total_income'] = data['total_income']

# –ø—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é categorize_income_2 –∫ —Å—Ç–æ–ª–±—Ü—É 'total_income' –Ω–∞—à–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
data_an['total_inc_cat_new'] = data_an['total_income'].apply(categorize_income_2)

# —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ª–∏—Ü —Å –Ω–æ–≤—ã–º–∏ —É—Ä–æ–≤–Ω—è–º–∏ –¥–æ—Ö–æ–¥–æ–≤, –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç–∏
pivot_income_new = data_an.pivot_table(index='total_inc_cat_new', columns='debt_category', values='debt', aggfunc='count')

# –ø–æ—Å—á–∏—Ç–∞–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç –¥–æ–ª–∂–Ω–∏–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–æ—Ö–æ–¥–∞, —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é –∫–æ–ª–æ–Ω–∫—É, –æ–∫—Ä—É–≥–ª–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é

pivot_income_new['debt %'] = pivot_income_new['debt'] / (pivot_income_new['debt'] + pivot_income_new['clear']) * 100
pivot_income_new['debt %'] = round(pivot_income_new['debt %'], 1)
pivot_income_new.sort_values('debt %', ascending=False)
```
