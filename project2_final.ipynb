```python
# выведем список уникальных значений столбца "locality_name" для последующего устранения неявных дубликатов
realty['locality_name'].unique()
```

   *Составим таблицу только топонимов, встречающихся в вышеприведенном уникальном списке 2 раза и более. Нас будут интересовать только названия, без статуса населенного пункта. Напишем функцию,создающую эту таблицу*

```python
def unic_topo(topo): 
    """
    Выводит на экран таблицу из 2-х колонок: 
    - список исключительно топонимов, без статуса населенного пункта;
    - сколько раз топоним повторяется в уникальном списке со статусами, минимум 2 повтора
    """
    # формируем список уникальных значений
    mylist = list(topo.unique()) 

    #создаём список из слов, от которых надо избавиться
    del_list = ['посёлок', 'городской', 'поселок', 'деревня', 'городского типа', 'садовое', 'садоводческое', 
                'некоммерческое', 'товарищество', 'село', 'имени', 'станции', 'при', 'железнодорожной', 'коттеджный', ' ']

    for i in range(len(mylist)): # берём каждый элемент большого списка
        mylist[i] = str(mylist[i]) # на всякий случай убедимся, что только строки
    
        for elem in del_list:     # каждый удаляемый элемент 
            if elem in mylist[i]:  # проверяем на наличие в элементе большого списка
                mylist[i] = mylist[i].replace(elem, '')  # если да, удаляем элемент

    spis = pd.Series(mylist).value_counts() # получившийся список преобразуем в серию датафрейма с подсчётом уникальных

    locality=[] # список для формирования топонимов
    mentions=[] # список для формирования числа уникальных, более 2-х

    for i in range(len(spis)): # проходим по серии датафрейма с подсчитанными уникальными
        if spis[i] > 1: # если число повторов более 1
        
            locality.append(spis.index[i]) # формируем список топонимов
            mentions.append(spis[i]) # формируем список повторов топонимов
            
    # формируем датафрейм топоним-повторы
    print(pd.DataFrame({'Нас.пункт': locality, 'Уник.упоминания': mentions }))
    
# запускаем функцию поиска топонимов-дубликатов
unic_topo(realty['locality_name'])    
```

  *Анализ самых распространённых топонимов по изначальному списку показывает, что часто они имеют наименования "посёлок" и "поселок", что, очевидно, одно и то же. Можно их объединить в одно.*
   
   *Аналогично много топонимов-двойников имеют статус как "городской посёлок", так и "посёлок городского типа". Проверка на Яндекс-картах названиий "городской посёлок" и "посёлок городского типа" показала их идентичность. Населенные пункты из списка с "посёлком городского типа" отображаются на картах со статусом "городской посёлок". Будем руководствоваться терминологией Яндекс-карт, заменим "посёлок городского типа" на "городской посёлок"*  

```python
# произведём замены следующих статусов населённых пунктов в датафрейме
realty['locality_name'] = realty['locality_name'].str.replace('поселок', 'посёлок') 
realty['locality_name'] = realty['locality_name'].str.replace('посёлок городского типа', 'городской посёлок')
```

   *Посмотрим, сколько осталось неявных дубликатов, запустив снова функцию поиска уникальных топонимов-дубликатов*

```python
# ищем уникальные топонимы-дубликаты после замен в столбце 'locality_name'
unic_topo(realty['locality_name'])
```

   *18 топонимов-двойников. Проверка по уникальному списку и Яндекс-картам показала, что почти все из них, кроме трёх, - это действительно разные населенные пункты, и объединять их нельзя. Например, село Павлово и городской посёлок Павлово находятся в разных районах области, между ними 40 км на автомобиле. Любань и посёлок Любань - разные образования, хотя и граничат на карте.     *посёлок Мурино - Мурино (примыкает к Санкт-Петербургу)
    *посёлок Гарболово - деревня Гарболово

```python
# произведём замены следующих статусов населённых пунктов в датафрейме
realty['locality_name'] = realty['locality_name'].str.replace('деревня Кудрово', 'Кудрово')
realty['locality_name'] = realty['locality_name'].str.replace('посёлок Мурино', 'Мурино') 
realty['locality_name'] = realty['locality_name'].str.replace('посёлок Гарболово', 'деревня Гарболово')
```

*Можно ещё посмотреть на список уникальных топонимов, после замен, отсортированный по алфавиту, нет ли грамматических ошибок типа "Зеленогорск/Зелиногорск". Так как мы не можем здесь обратиться к уже создававшемуся такому списку, поскольку переменная внутри функции, то снова возьмём уникальные населёнки из датасета и очистим их от статусов.*

```python
# список уникальных топонимов из датасета
list_unic = list(realty['locality_name'].unique()) 

#создаём список из слов, от которых надо избавиться
del_list = ['посёлок', 'городской', 'поселок', 'деревня', 'городского типа', 'садовое', 'садоводческое', 
                'некоммерческое', 'товарищество', 'село', 'имени', 'станции', 'при', 'железнодорожной', 'коттеджный', ' ']

for i in range(len(list_unic)): # берём каждый элемент большого списка
    
    for elem in del_list:     # каждый удаляемый элемент 
        if elem in list_unic[i]:  # проверяем на наличие в элементе большого списка
            list_unic[i] = list_unic[i].replace(elem, '')  # если да, удаляем элемент
            
list(pd.Series(list_unic).sort_values()) # выводим список топонимов, отсортированный по алфавиту, в удобном виде
```
