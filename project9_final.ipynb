## –£–º–Ω–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü

–í —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏ –≤—ã –º–æ–∂–µ—Ç–µ –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å —Ñ–æ—Ä–º—É–ª—ã –≤ *Jupyter Notebook.*

–ß—Ç–æ–±—ã –∑–∞–ø–∏—Å–∞—Ç—å —Ñ–æ—Ä–º—É–ª—É –≤–Ω—É—Ç—Ä–∏ —Ç–µ–∫—Å—Ç–∞, –æ–∫—Ä—É–∂–∏—Ç–µ –µ—ë —Å–∏–º–≤–æ–ª–∞–º–∏ –¥–æ–ª–ª–∞—Ä–∞ \\$; –µ—Å–ª–∏ —Å–Ω–∞—Ä—É–∂–∏ ‚Äî  –¥–≤–æ–π–Ω—ã–º–∏ —Å–∏–º–≤–æ–ª–∞–º–∏ \\$\\$. –≠—Ç–∏ —Ñ–æ—Ä–º—É–ª—ã –∑–∞–ø–∏—Å—ã–≤–∞—é—Ç—Å—è –Ω–∞ —è–∑—ã–∫–µ –≤—ë—Ä—Å—Ç–∫–∏ *LaTeX.* 

–î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –º—ã –∑–∞–ø–∏—Å–∞–ª–∏ —Ñ–æ—Ä–º—É–ª—ã –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ú–æ–∂–µ—Ç–µ –∏—Ö —Å–∫–æ–ø–∏—Ä–æ–≤–∞—Ç—å –∏ –æ—Ç—Ä–µ–¥–∞–∫—Ç–∏—Ä–æ–≤–∞—Ç—å, —á—Ç–æ–±—ã —Ä–µ—à–∏—Ç—å –∑–∞–¥–∞—á—É.

–†–∞–±–æ—Ç–∞—Ç—å –≤ *LaTeX* –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ.

–û–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:

- $X$ ‚Äî –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–Ω—É–ª–µ–≤–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –µ–¥–∏–Ω–∏—Ü)

- $y$ ‚Äî –≤–µ–∫—Ç–æ—Ä —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞

- $P$ ‚Äî –º–∞—Ç—Ä–∏—Ü–∞, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é —É–º–Ω–æ–∂–∞—é—Ç—Å—è –ø—Ä–∏–∑–Ω–∞–∫–∏

- $w$ ‚Äî –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–Ω—É–ª–µ–≤–æ–π —ç–ª–µ–º–µ–Ω—Ç —Ä–∞–≤–µ–Ω —Å–¥–≤–∏–≥—É)

–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è:

$$
a = Xw
$$

–ó–∞–¥–∞—á–∞ –æ–±—É—á–µ–Ω–∏—è:

$$
w = \arg\min_w MSE(Xw, y)
$$

–§–æ—Ä–º—É–ª–∞ –æ–±—É—á–µ–Ω–∏—è:

$$
w = (X^T X)^{-1} X^T y
$$

```python
# —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫

features = ins.drop(columns='benefits')
target = ins['benefits']
print(features.shape)
print(target.shape)
```

*–î–ª—è –±–æ–ª—å—à–µ–π –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –ø—Ä–∏ –ø–æ—Å–ª–µ–¥—É—é—â–µ–π –ø—Ä–æ–≤–µ—Ä–∫–µ —Ñ–æ—Ä–º—É–ª –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –≤ —ç—Ç–æ–º —Ä–∞–∑–¥–µ–ª–µ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å **–≤–µ—Å—å** –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –æ–±—É—á–µ–Ω–∏—è-–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –±–µ–∑ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â—É—é-—Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏.*

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –ø–æ —Ñ–æ—Ä–º—É–ª–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

class LinearRegressionFormula:
    def __init__(self, train_features, train_target):
        
        # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ –∏–∑ –µ–¥–∏–Ω–∏—Ü –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ —Å–¥–≤–∏–≥–∞ w0
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1) 
        y = train_target
        w = np.linalg.inv(X.T @ X) @ X.T @ y # —Ä–∞—Å—á—ë—Ç –≤–µ–∫—Ç–æ—Ä–∞ –≤–µ—Å–æ–≤ –∏ —Å–¥–≤–∏–≥–∞
        self.w = w[1:] # –≤–µ—Å–∞
        self.w0 = w[0] # —Å–¥–≤–∏–≥
        
    # —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
    def predict(self, test_features):
        return test_features.dot(self.w) + self.w0 
    
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∞—Å—Å–∞ –ø–æ –∏—Å—Ö–æ–¥–Ω—ã–º –¥–∞–Ω–Ω—ã–º, —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –º–µ—Ç—Ä–∏–∫–∞
model = LinearRegressionFormula(features, target)
predictions = model.predict(features)
print(r2_score(target, predictions))
```

```python
# —Å–¥–≤–∏–≥ –ø–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
model.w0
```

```python
# –≥–µ–Ω–µ—Ä–∞—Ü–∏—è –º–∞—Ç—Ä–∏—Ü—ã-–∫–ª—é—á–∞, —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –Ω–∞ –æ–±—Ä–∞—Ç–∏–º–æ—Å—Ç—å

while 1>0:
    P = np.random.randint(1, 101, size=(features.shape[1], features.shape[1]))
    
    try:
        np.linalg.inv(P)
        break
        
    except:
        continue
print(P)
```

```python
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∞—Å—Å–∞ –ø–æ —à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º, —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –º–µ—Ç—Ä–∏–∫–∞

model_2 = LinearRegressionFormula(features @ P, target)
predictions_2 = model_2.predict(features @ P)
print(r2_score(target, predictions_2))
```

```python
# —Å—É–º–º–∞—Ä–Ω–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –ø–æ –∏—Å—Ö–æ–¥–Ω—ã–º –∏ —à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º
(predictions_2 - predictions).abs().sum()
```

```python
# –≤–µ—Å–∞ –º–æ–¥–µ–ª–∏ –ø–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º —à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
model_2.w
```

```python
# —Å–¥–≤–∏–≥ –ø–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º —à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
model_2.w0
```

**–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ:** 

–ü–æ—Å–∫–æ–ª—å–∫—É –º—ã –¥–æ–º–Ω–æ–∂–∞–µ–º –∏—Å—Ö–æ–¥–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É $X$ –Ω–∞ –º–∞—Ç—Ä–∏—Ü—É $P$, —Ç–æ –Ω–æ–≤—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –±—É–¥—É—Ç –≤—ã—Ä–∞–∂–∞—Ç—å—Å—è –∫–∞–∫ $a' = XP w'$.

–ò—Å—Ö–æ–¥–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è: $a = X w$ . –î–æ–∫–∞–∂–µ–º, —á—Ç–æ  –Ω–æ–≤—ã–µ –∏ –∏—Å—Ö–æ–¥–Ω—ã–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è —Ä–∞–≤–Ω—ã, —Ç–æ –µ—Å—Ç—å $XP w' = X w$.


–§–æ—Ä–º—É–ª–∞ –æ–±—É—á–µ–Ω–∏—è –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: $w = (X^T X)^{-1} X^T y$. 

–§–æ—Ä–º—É–ª–∞ –æ–±—É—á–µ–Ω–∏—è –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö (—É–º–Ω–æ–∂–∞–µ–º $X$ –Ω–∞ $P$): $w' = ((X P)^T (X P))^{-1} (X P)^T y$. 

–£–ø—Ä–æ—Å—Ç–∏–º –ø–æ—Å–ª–µ–¥–Ω–µ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ.

** **


$w' = (P^T X^T X P)^{-1} (P^T X^T) y$, —Ç–∞–∫ –∫–∞–∫ –ø–æ —Å–≤–æ–π—Å—Ç–≤—É —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã $(A B)^T = B^T A^T$

$w' = (P^T (X^T X) P)^{-1} (P^T X^T) y$, –≤—ã–¥–µ–ª—è–µ–º 3 –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ –º–∞—Ç—Ä–∏—Ü—ã: $P^T$, $X^T X$ –∏ $P$

$w' = P^{-1} (X^T X)^{-1} (P^T)^{-1} P^T X^T y$, —Ç–∞–∫ –∫–∞–∫ $(A B C)^{-1} = (A (B C))^{-1} = (B C)^{-1} A^{-1} = C^{-1} B^{-1} A^{-1}$, —Å–≤–æ–π—Å—Ç–≤–æ –ø—Ä–∏–º–µ–Ω–∏–º–æ —Ç–æ–ª—å–∫–æ –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü

$w' = P^{-1} (X^T X)^{-1} E X^T y$, —Ç–∞–∫ –∫–∞–∫ $A^{-1} A = E$ , –∞ $P^T$ - –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞

$w' = P^{-1} (X^T X)^{-1} X^T y$, —Ç–∞–∫ –∫–∞–∫ $A E = A$, –ø—Ä–∏ —ç—Ç–æ–º —É $E$ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø–æ —á–∏—Å–ª—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–∏–∑-–∑–∞ –º–∞—Ç—Ä–∏—Ü  $P^T$ –∏ $(P^T)^{-1}$), –∞ $(X^T X)^{-1}$ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –∏ —Ç–∞–∫–∂–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é –≤ —á–∏—Å–ª–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

** **

–ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º $w$ –∏ $w'$ –≤ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º–æ–µ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ $XP w' = X w$:

$X P P^{-1} (X^T X)^{-1} X^T y = X (X^T X)^{-1} X^T y$

$X E (X^T X)^{-1} X^T y = X (X^T X)^{-1} X^T y$, —Ç–∞–∫ –∫–∞–∫ $A A^{-1} = E$, –∞ $P$ - –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞

$X (X^T X)^{-1} X^T y = X (X^T X)^{-1} X^T y$, —Ç–∞–∫ –∫–∞–∫ $E A = A$, –ø—Ä–∏ —ç—Ç–æ–º —É $E$ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø–æ —á–∏—Å–ª—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–∏–∑-–∑–∞ –º–∞—Ç—Ä–∏—Ü  $P$ –∏ $P^{-1}$), –∞ $(X^T X)^{-1}$ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –∏ —Ç–∞–∫–∂–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é –≤ —á–∏—Å–ª–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

$X (X^T X)^{-1} X^T y = X (X^T X)^{-1} X^T y$

–ü–æ–ª—É—á–∏–ª–∏ —è–≤–Ω–æ–µ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–æ –∏ –ø–æ—Å–ª–µ —É–º–Ω–æ–∂–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü —Ä–∞–≤–Ω—ã, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –º–µ—Ç—Ä–∏–∫–∏ –ø–æ –∏—Å—Ö–æ–¥–Ω—ã–º –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º —Ç–æ–∂–µ —Ä–∞–≤–Ω—ã.

** **

*6. –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∏ –Ω–µ–∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞ —Ä–∞–≤–Ω—ã ($XP w' = X w$), —ç—Ç–æ –¥–æ–∫–∞–∑–∞–Ω–æ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º —Ä–∞–∑–¥–µ–ª–µ. –í–µ–∫—Ç–æ—Ä —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤—Å–µ–≥–¥–∞ –Ω–µ–∏–∑–º–µ–Ω–µ–Ω, –∑–Ω–∞—á–∏—Ç –∏ –º–µ—Ç—Ä–∏–∫–∞ R2 (–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏), –∑–∞–≤–∏—Å—è—â–∞—è –æ—Ç –∏—Å—Ö–æ–¥–Ω—ã—Ö –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –±—É–¥–µ—Ç –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –≤ –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–º –∏ –Ω–µ–∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–∞—Ö.*

```python
# –≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã-–∫–ª—é—á–∞, —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –Ω–∞ –æ–±—Ä–∞—Ç–∏–º–æ—Å—Ç—å

while 1>0:
    P = np.random.randint(1, 101, size=(features.shape[1], features.shape[1]))
    
    if np.linalg.det(P) == 0:
        continue
        
    else:
        break
print(P)
```
